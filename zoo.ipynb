```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("Zoo.csv")
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.rename(columns={"animal name":"animalname"},inplace=True)
```

```python
df.head(10)
```

```python
#Dummies
dummies=pd.get_dummies(df["animalname"])
dummies
```

```python
df1=pd.concat([df,dummies],axis=1)
df1=df1.drop(["animalname"],axis=1)
df1
```

```python
labels=df1.iloc[:,16]
labels
```

```python
features=df1.drop(["type"],axis=1)
features
```

```python
#Normalize the Functio
def norm_func(i):
    x=(i-i.min()/i.max()-i.min())
    return(x)
```

```python
data=norm_func(features)
data
```

```python
#Split model
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(data,labels,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KN
model=KN(n_neighbors=5)
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```

```python
#Train Accuracy
Train_acc=np.mean(model.predict(X_train)==Y_train)
Train_acc
```

```python
#Test Accuracy
Test_acc=np.mean(model.predict(X_test)==Y_test)
Test_acc
```

```python
model1=KN(n_neighbors=5)
model1.fit(X_train,Y_train)
```

```python
acc = []
# Calculating error for K values between 1 and 40
for i in range(1,30):
    model1=KN(n_neighbors=i)
    model1.fit(X_train,Y_train)
    pred_i = model1.predict(X_test)
    acc.append(np.mean(pred_i != Y_test))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1,30), acc, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```
